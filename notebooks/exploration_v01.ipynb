```python
import os
import sys

import nb_config

import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
```

```python
env = nb_config.get_root_path()
dataset_path = f'{env}/dataset'
eg_img_path = f'{env}/dataset/rain/14.jpg'
```

```python
def load_dataset(path: str):
    # Load dataset from path and resize images to 128x128
    dataset = keras.utils.image_dataset_from_directory(path, labels="inferred", batch_size=32, image_size=(128, 128))
    return dataset
```

```python
def load_image(path: str):
    img = keras.utils.load_img(path, target_size=(128, 128))
    return img
```

```python
dataset = load_dataset(dataset_path)
```

```python
image = load_image(eg_img_path)
```
